```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 11000000
k = [0]*n
k2 = [0]*n
j = [0]*n
j[1] = 1
k[1] = 1
k2[1] = 1
for i in range(2,n):
    j[i] = i
    k[i] = k[i - 1] + 1/i
    k2[i] = k2[i - 1] + 1/(i*i)
sums = pd.DataFrame({'n':j, 'sum{1/k}':k, 'sum{1/k^2}':k2})
sums.tail(50)
```
